### 진짜 내 생일의 기온 변화를 그래프로 그리기

```python
import csv
import matplotlib.pyplot as plt
```

```python
f = open('seoul.csv')
data = csv.reader(f)
```

```python
# header
next(data)
```

```python
# variable 
high = []
low  = []
```

```python
# Logic
for row in data :
    if row[-1] != '' and row[-2] != '' :  # 최고 기온 값과 최저 기온 값이 공백이 아니라면
        date = row[0].split('-')           # 날짜 값을 -로 구분하여 date 리스트에 저장
        
        if 1983 <= int(date[0]):            # date 리스트의 첫번째 인덱스 값 (년도)를 int로 casting
                                            # 1983년 이후의 데이터들만
            if date[1] == '12' and date[2] == '25': # 12월 25일 이라면
                
                high.append(float(row[-1])) # 최고 기온 값을 high에 저장
                low.append(float(row[-2])) # 최저 기온 값을 low에 저장
                
plt.rc('font', family='Malgun Gothic') # 맑은 고딕을 기본 글꼴로 지정
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지
plt.title('내 생일의 기온 변화 그래프') # 제목 설정
plt.plot(high, 'hotpink', label='high')
plt.plot(low, 'skyblue', label='low')
plt.legend() # 범례 설정
plt.show()   # 그래프 표시
```
