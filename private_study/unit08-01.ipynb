```python
import csv
```

```python
f = open('age.csv')
data = csv.reader(f)
result = []
```

```python
name = input("원하는 읍면동을 입력하세요 : ")
```

```python
for row in data :
    if name in row[0] :
        for i in row[3:] :
            result.append(int(i.replace(',','')))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.style.use('ggplot')
plt.rc('font', family='Malgun Gothic')
plt.title(name+' 지역의 인구구조')
plt.plot(result)
plt.show()
```
