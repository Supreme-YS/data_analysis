```python
import matplotlib.pyplot as plt
import csv
```

```python
f = open('seoul.csv')
data = csv.reader(f)
next(data)
```

```python
month = [[],[],[],[],[],[],[],[],[],[],[],[]]
```

```python
for row in data :
    if row[-1] != '' :
        month[int(row[0].split('-')[1])-1].append(float(row[-1]))
```

```python
plt.boxplot(month)
plt.show()
```
