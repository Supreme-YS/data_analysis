```python
for i in tqdm(range(len(f_list))):
    datadf = pd.read_csv(f_list[i], error_bad_lines=False, encoding='UTF-8')
    datadf = datadf.groupby(by=['id', 'x', 'y', 'timezn_cd']).sum().reset_index().iloc[:,:-2]
    datadf = datadf[datadf.id > 0]
    
    datadf['m'] = datadf.iloc[:,4:18].sum(axis=1)
    datadf['f'] = datadf.iloc[:,18:32].sum(axis=1)
    
    col1 = 'm_'+str(f_list[i][31:37])
    col2 = 'f_'+str(f_list[i][31:37])
    
    datadf = datadf.rename(columns = {'m':col1,'f':col2})
    a = datadf[['id','x','y','timezn_cd', col1, col2]]
    if i == 0:
        fpdf = a
    else:
        fpdf = pd.merge(fpdf, a, how='outer', on=['id','x','y','timezn_cd'])
    
    print(f_list[i])
    
fpdf.to_csv('dataset_fp.csv', encoding = "UTF-8" ,index = False)
print("DONE :)")
```
