```python
import pandas as pd
import numpy as np
import os
from tqdm import tqdm
```

```python
file_list = os.listdir("./")
f_list = [file for file in file_list if file.endswith(".csv")]
f_list
```

```python
datadf = pd.read_csv('100c_201901_busan_dense.csv', encoding='UTF-8')
datadf = datadf.groupby(by=['id', 'x', 'y']).sum().reset_index()
datadf = datadf[datadf.id > 0]

datadf.iloc[:,3:]
```

```python
for i in tqdm(range(len(f_list))):
    datadf = pd.read_csv(f_list[i], error_bad_lines=False, encoding='UTF-8')
    datadf = datadf.groupby(by=['id', 'x', 'y']).sum().reset_index()
    datadf = datadf[datadf.id > 0]
    
    datadf['new_total'] = datadf.iloc[:,3:].sum(axis=1)
    
    col1 = 'new_total'+str(f_list[i][5:11])

    datadf = datadf.rename(columns = {'new_total':col1})
    a = datadf[['id','x','y', col1]]
    if i == 0:
        fpdf = a
    else:
        fpdf = pd.merge(fpdf, a, how='outer', on=['id','x','y'])
    
    print(f_list[i])
    
fpdf.to_csv('100c_busan_dense_total.csv', encoding = "UTF-8" ,index = False)
print("DONE :)")
```
