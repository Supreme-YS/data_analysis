```python
import pandas as pd
import numpy as np

import glob

print('>>> 본 소스 작성 시 사용된 라이브러리 버전에 유의하세요!')
f'pd : {pd.__version__}  |  np : {np.__version__}'
```

## 개별 테이블 로드 및 내용 확인

```python
dong_df = pd.read_pickle('서울특별시_월별_행정동별_총합.pkl')
dong_df.columns = ['시군구코드', '시군구명', '행정동코드', '행정동명', '연도별', '월별', '유동인구총합']
dong_df
```

```python
time_df = pd.read_pickle('서울특별시_월별_행정동별_시간대별_총합.pkl')
time_df.columns = ['시군구코드', '시군구명', '행정동코드', '행정동명', '연도별', '월별', '시간대별', '유동인구총합']
time_df
```

```python
sex_age_df = pd.read_pickle('서울특별시_월별_행정동별_성연령별_총합.pkl')
sex_age_df['sex'] = sex_age_df.sex_age.str[0].str.upper()
sex_age_df['age'] = sex_age_df.sex_age.str[1:].astype(int)
sex_age_df = sex_age_df.drop(columns=['sex_age'])[['sgg_cd', 'sgg_nm', 'admdong_cd', 'admdong_nm',
                                                   'year', 'month', 'sex', 'age', 'total']]
sex_age_df.columns = ['시군구코드', '시군구명', '행정동코드', '행정동명', '연도별', '월별', '성별', '연령대별', '유동인구총합']
sex_age_df
```

## 월별 자치구별 전체, 월별 서울특별시 전체 테이블 생성

```python
gu_df = dong_df.drop(columns=['행정동코드', '행정동명']).\
            groupby(['시군구코드', '시군구명', '연도별', '월별'])['유동인구총합'].sum().reset_index()
gu_df
```

```python
seoul_df = gu_df.drop(columns=['시군구코드', '시군구명']).\
            groupby(['연도별', '월별'])['유동인구총합'].sum().reset_index()
seoul_df
```

## 최종 엑셀 파일 저장 및 확인

```python
with pd.ExcelWriter('서울특별시_유동인구_집계_최종.xlsx') as writer:
    seoul_df.to_excel(writer, sheet_name='월별 서울특별시 유동인구', index=False)
    gu_df.to_excel(writer, sheet_name='월별 자치구별 유동인구', index=False)
    dong_df.to_excel(writer, sheet_name='월별 행정동별 유동인구', index=False)
    sex_age_df.to_excel(writer, sheet_name='월별 성연령별 유동인구', index=False)
    time_df.to_excel(writer, sheet_name='월별 시간대별 유동인구', index=False)

# 파일 저장 확인
glob.glob('서울특별시_유동인구_집계_최종.xlsx')
```
