```python
import pandas as pd
import numpy as np
import os
from tqdm import tqdm
```

```python
path = "C:/Users/dudtj/busan_moving_person/data/opt_data"
```

```python
file_list = os.listdir(path)
os.chdir(path)
f_list = [file for file in file_list if file.endswith(".csv")]
f_list
```

```python
optDF = pd.read_csv(f_list[0], encoding="UTF-8")
optDF.iloc[:,4:28:2]
optDF.iloc[:,5:28:2]
optDF.head(5)
```

```python
for i in tqdm(range(len(f_list))):
    datadf = pd.read_csv(f_list[i], encoding='UTF-8')
    datadf = datadf.groupby(by=['id', 'x', 'y']).sum().reset_index()
    datadf = datadf[datadf.id > 0]
    
    datadf['m'] = datadf.iloc[:,4:28:2].sum(axis=1)
    datadf['f'] = datadf.iloc[:,5:28:2].sum(axis=1)
    
    col1 = 'm'
    col2 = 'f'
    
    a = datadf[['id','x','y',col1, col2]]
    if i == 0:
        fpdf = a
    else:
        fpdf = pd.merge(fpdf, a, how='outer', on=['id','x','y'])
    
    print(f_list[i])
    
fpdf.to_csv('opt_analysis.csv', encoding = "UTF-8" ,index = False)
print("DONE :)")
```
